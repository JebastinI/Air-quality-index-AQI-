## DATA IMPUTATION

* Null values in the Data is imputed using KNN-imputer.
* KNN-imputer => KNNimputer is a scikit-learn class used to fill out or predict the missing values in a dataset.It is a more useful method which works on the basic approach of the KNN algorithm rather than the naive approach of filling all the values with mean or the median. In this approach, we specify a distance from the missing valueswhich is also known as the K parameter.The missing value will be imputed in reference to the mean of the neighbours.

* Some of the stations have no records of the particullar pollutant so we impute the missing values with zero for better prediction.

```python
# getting the pollutant names for the station where there are no records.
null_sid=[]
dic={}
for i in station_id:
    for j in col:
        if ( ~  (df[df["StationId"]==i][j].isnull().sum() <  len(df[df["StationId"]==i][j]))  ):
            null_sid.append({i:j})
null_sid        
```

```python
len(null_sid)
```

```python
# getting the pollutant names for the station where there are less than 5 records to find out the neighours for the pollutant.

null_sid1=[]
#dic={}
for i in station_id:
    for j in col:
        if ( ~ (df[df["StationId"]==i][j].notnull().sum()>=5)  ):
            null_sid1.append({i:j})
null_sid1 
```

```python
len(null_sid1)
```

```python
# getting the name of the pollutant have less than 5 non-null values. 
for i in null_sid1:
    if i in null_sid:
        continue
    else:
        print(i)
        null_sid.append(i)
```

```python
# imputing the missing values with zero where there is no records of the pollutant in particular station.
for i in station_id:
    for j in col:
        if ( ~  (df[df["StationId"]==i][j].isnull().sum() <  len(df[df["StationId"]==i][j]))  ):
            df.loc[((df.StationId == i)  & (df[j].isnull())) , j] = 0
```

```python
for i in null_sid:
    for j in col:
        print(df[df["StationId"]==i][j].isnull().sum())
```

```python
# for Benezene feature in DL006 station we have two records, so we impute that  with mean of the two values. 
df.loc[((df.StationId == "DL006")  & (df["Benzene"].isnull())) , "Benzene"] = 1.1
```

```python
df[ df["StationId"]== "DL006" ]["Benzene"].isnull().sum()
```

```python
col=['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO',
       'SO2', 'O3', 'Benzene', 'Toluene']
```

```python
# KNNImputer for the rest of the null values.

from sklearn.impute import KNNImputer
for k in col:
    l=[]
    for i in station_id:
        
        imputer = KNNImputer(n_neighbors=5)
        temp=df[df["StationId"]==i]
        temp_df=pd.DataFrame(imputer.fit_transform(temp[[k]]))
        for j in temp_df[0]:
            l.append(j)
    df[k]=l
```

### STATISTICAL TEST FOR FINDING THE SIGNIFICANT AND INSIGNIFICANT VARIABLES

```python
import scipy.stats as stats

def t_test1(col,tgt):
      sig_var =[]
      insig_var =[]

      for i in col:
        # print(i)
        # print(len(i),len(tgt))
        test_stat,p_val=stats.pearsonr(df[i],tgt)
        if p_val < 0.005 :
          sig_var.append(i)
        else:
          insig_var.append(i)
      return sig_var , insig_var

si,ins=t_test1(col,df["AQI"])

print("THE SIGNIFICANT VARIABLES ARE : ",si)
print("THE INSIGNIFICANT VARIABLES ARE : ",ins)
```

### Heteroscedasticity

* variance at the first will not be same at the last ,eg: income vs expense when income increase the expense increases then the variance is also increased.
* unequal variance
* it is guaged as the data which has the variance not equal at different points
* the shape of the data is either open funnel or closed funnel
* thus it gives a pattern of variance getting distorted at different levels.
* the heteroscedasticity is present in the data due to the presence of the outliers. one can try to remove heteroscedasticity by removing the outliers or applying the log transformation.


test=> BreuschPagan Test

ho: equal variance present in the model

h1: false

```python
# breuschpagan test

import statsmodels.stats.api as ssa


# model.exog: it represents all the data points/ input variables

ssa.het_breuschpagan(model.resid, model.model.exog)

# if p value is < 0.05 we reject is Ho
```

### DECISION TREE REGRESSOR

```python
# DECISION TREE REGRESSOR
from sklearn.model_selection import KFold
kfold=KFold(n_splits=5,shuffle=True,random_state=42)
from sklearn.tree import DecisionTreeRegressor

dt=DecisionTreeRegressor(random_state=10)

rmse_train=[]
rmse_test=[]

for train_index,test_index in kfold.split(x,y):
    xtrain=x.iloc[train_index]
    ytrain=y.iloc[train_index]
    xtest=x.iloc[test_index]
    ytest=y.iloc[test_index]
    pred_train=dt.fit(xtrain,ytrain).predict(xtrain)
    pred_test=dt.fit(xtrain,ytrain).predict(xtest)
    rmse_train.append(np.sqrt(mean_squared_error(ytrain,pred_train)))
    rmse_test.append(np.sqrt(mean_squared_error(ytest,pred_test)))
    
print(np.mean(rmse_train))
print(np.mean(rmse_test))
```

```python
model_performance = pd.DataFrame()
model_performance["Model"]=pd.Series(["DecisionTree"])
model_performance["Train_rmse"]=pd.Series(np.mean(rmse_train))
model_performance["Test_rmse"]=pd.Series(np.mean(rmse_test))
```

```python
model_performance
```

## RIDGE MODEL

```python
# RIDGE MODEL
from sklearn.model_selection import KFold
kfold=KFold(n_splits=5,shuffle=True,random_state=42)
from sklearn.linear_model import Ridge

r=Ridge(alpha=0.5)

rmse_train=[]
rmse_test=[]

for train_index,test_index in kfold.split(x,y):
    xtrain=x.iloc[train_index]
    ytrain=y.iloc[train_index]
    xtest=x.iloc[test_index]
    ytest=y.iloc[test_index]
    pred_train=r.fit(xtrain,ytrain).predict(xtrain)
    pred_test=r.fit(xtrain,ytrain).predict(xtest)
    rmse_train.append(np.sqrt(mean_squared_error(ytrain,pred_train)))
    rmse_test.append(np.sqrt(mean_squared_error(ytest,pred_test)))
    
print(np.mean(rmse_train))
print(np.mean(rmse_test))
```

```python
dict1={"Model":"Ridge Model",
       "Train_rmse":np.mean(rmse_train),
       "Test_rmse":np.mean(rmse_test)
}
model_performance=model_performance.append(dict1,ignore_index=True)
```
